### The next steps are comparing loudness of two audio recordings, one where listener 1 speaks at the end and the other

```python
import soundfile as sf
import pyloudnorm as pyln

data, rate = sf.read("Data/Audio/Osi Audio Extracted.wav") # load audio

# peak normalize audio to -1 dB
peak_normalized_audio = pyln.normalize.peak(data, -1.0)

# measure the loudness first 
meter = pyln.Meter(rate) # create BS.1770 meter
loudness = meter.integrated_loudness(data)

# loudness normalize audio to -12 dB LUFS
listener_then_speaker_loudness_normalized_audio = pyln.normalize.loudness(data, loudness, -20.0)
```

```python
import soundfile as sf
import pyloudnorm as pyln

data, rate = sf.read("Data/Audio/Scott Audio Extracted.wav") # load audio

# peak normalize audio to -1 dB
peak_normalized_audio = pyln.normalize.peak(data, -1.0)

# measure the loudness first 
meter = pyln.Meter(rate) # create BS.1770 meter
loudness = meter.integrated_loudness(data)

# loudness normalize audio to -12 dB LUFS
speaker_to_listener_loudness_normalized_audio = pyln.normalize.loudness(data, loudness, -20.0)
```

```python
speaker_to_listener_df=pd.DataFrame(speaker_to_listener_loudness_normalized_audio[5856000:7680000])
speaker_to_listener_df.plot()
```
